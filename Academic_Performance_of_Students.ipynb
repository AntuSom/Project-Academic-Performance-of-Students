```python
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Importing_CSV
Academic_Performance = pd.read_csv('/content/Academic Performance of Students.csv')
```

```python
#Displaying data set
Academic_Performance.head(-1)
```

```python
#Function assigning Grades
def assign_grade(score):
  if score >=90:
    return 'A'
  elif score >=80:
    return 'B'
  else:
    return 'C'
```

```python
#Assigning grades as per function
Academic_Performance['Grade']=Academic_Performance['Final_Score'].apply(assign_grade)
```

```python
Academic_Performance.head()
```

```python
#Filtered Students who perticipated in Extracurricular_Activities
Academic_Performance.loc[Academic_Performance['Extracurricular_Activities']=='Yes']
```

```python
#Group by Parental_Education_Level & Final_Score
FinalScore=Academic_Performance.groupby('Parental_Education_Level')['Final_Score'].mean()
print(FinalScore)
```

```python
#Corelation between Study_Hours,Attendance_Percentage,Final_Score
Corelation =Academic_Performance [['Study_Hours','Attendance_Percentage','Final_Score']].corr()
print(Corelation)
#Plotting heatmap
sns.heatmap(Corelation, annot=True)
plt.show()
```

```python
#Corelation between SAT_Score,Attendance_Percentage

Sat_att=Academic_Performance[['SAT_Score','Attendance_Percentage']].corr()
print(Sat_att)
```

```python
#Gender wise comparison of final score & bar chart

GENDER_SCORE=Academic_Performance.groupby('Gender')['Final_Score'].sum()
print(GENDER_SCORE)
plt.bar(GENDER_SCORE.index,GENDER_SCORE.values, color='blue',)
plt.xlabel('Gender')
plt.ylabel('Final Score')
plt.title('Gender vs Final Score')
plt.show()
```

```python
#Visulization of study_hours using density and histrogram

sns.kdeplot(Academic_Performance['Study_Hours'])
plt.show()
sns.distplot(Academic_Performance['Study_Hours'])
plt.show()

```
